```python
import pandas as pd
import numpy as np
from mplsoccer.pitch import Pitch
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C://Users//User//Desktop//csv codes//data//valladolidA.csv')
```

```python
df
```

```python
df = df[df['teamId'] == 'Barcelona']
df
```

```python
df['passer'] = df['playerId']
df['recipient'] = df['playerId'].shift(-1)
df
```

```python
passes = df[df['type'] == 'Pass']
passes
```

```python
successfullPasses = passes[passes['outcome'] == 'Successful']
successfullPasses
```

```python
subs = df[df['type'] == 'SubstitutionOff']
mins = subs['minute']
firstSub = mins.min()
```

```python
firstSub
```

```python
successfullPasses = successfullPasses[successfullPasses['minute'] < firstSub]
successfullPasses
```

```python
pas = pd.to_numeric(successfullPasses['passer'], downcast = 'integer')
rec = pd.to_numeric(successfullPasses['recipient'], downcast = 'integer')

successfullPasses['passer'] = pas
successfullPasses['recipient'] = rec
successfullPasses
```

```python
avg_location = successfullPasses.groupby('passer').agg({'x': ['mean'], 'y': ['mean', 'count']})
avg_location.columns = ['x', 'y', 'count']
```

```python
avg_location
```

```python
pass_bw = successfullPasses.groupby(['passer', 'recipient']).id.count().reset_index()
pass_bw.rename({'id': 'pass_count'}, axis = 'columns', inplace = True)

pass_bw = pass_bw.merge(avg_location, left_on = 'passer', right_index = True)
pass_bw = pass_bw.merge(avg_location, left_on = 'recipient', right_index = True, suffixes = ['', '_end'])
pass_bw
```

```python
pass_bw = pass_bw[pass_bw['pass_count'] > 3]
pass_bw
```

```python
pitch = Pitch(pitch_type = 'statsbomb', half = False,  pitch_color = 'grass', 
             line_color = 'white', stripe = True, corner_arcs = True)
fig, ax = pitch.draw(figsize = (16, 10))

arrows = pitch.arrows(1.2 * pass_bw.x, 0.8 * pass_bw.y, 1.2 * pass_bw.x_end, 0.8 * pass_bw.y_end, ax = ax, 
                     width = 2, headwidth = 3, zorder = 1, alpha = 0.5)
nodes = pitch.scatter(1.2 * avg_location.x, 0.8 * avg_location.y, color = 'red', ax = ax, 
                     zorder = 1, alpha = 1, s = 100)
```
